```python
import pandas as pd
import numpy as np
```

```python
# Generamos datos para los dataframe
cant_locacion = np.arange(10)
cant_categoria = np.arange(20)
cant_espectadores = np.arange(100)
capacidad = np.arange(100)
capacidad_extendida = np.arange(20)

fecha_inicio = '2018-09-25'
fecha_fin = '2018-11-01'

nombre_eventos = ["A", "B", "C","D", "E", "F", "G"]
nombre_sede = ["av","puerto_madero","tecnopolis"]
nombre_categoria = ["boxeo","natacion","voley","futbo","tenis","equitacion","yudo","remo","arco","dance"]
                    
                    
                    
eventos = pd.DataFrame({
    'fecha': pd.date_range(start=fecha_inicio, end=fecha_fin, periods=2000),
    'id_eventos': np.arange(2000),
    'id_locacion': np.random.choice(cant_locacion, 2000, replace=True),
    'id_categorias':np.random.choice(cant_categoria, 2000, replace=True),
    'nombre_evento': np.random.choice(nombre_eventos, 2000),
    'cant_espectadores': np.random.choice(cant_espectadores, 2000, replace=True),
})

locacion = pd.DataFrame({
    'id_locacion': np.arange(10),
   'nombre_locacion':["rojo","blanco","verde","amarillo","violeta","azul","gris","negro","naranja","celeste"],
   'capacidad': np.random.choice(capacidad, 10, replace=True),
    'capacidad_extendida': np.random.choice(capacidad_extendida, 10 , replace=True),
    'sede': np.random.choice(nombre_sede,10),
})


categorias_deportivas = pd.DataFrame({
    'id_categorias': np.arange(10),
    'nombre_categoria' : ["boxeo","natacion","voley","futbo","tenis","equitacion","yudo","remo","arco","dance"]
})
```

#### a) Nombre de la sede que acumulo la mayor cantidad de espectadores en eventos durante el certamen del 14 al 15 de octubre inclusive . Esto es de vial importancia para distribuir el merchandising oficial del evento para las fechas finales

Pasos a seguir:
* filtrar fecha de archivo evento
* inner join entre dataframe sede y eventos
* agrupo sede y cant_espectadores y saco el de mayor cant_espectadores

```python
eventos['fecha'] = pd.to_datetime(eventos['fecha'])
eventos['mes'] = eventos['fecha'].dt.month
eventos['dias']= eventos['fecha'].dt.day
eventos.mes.value_counts()
```

```python
#filtro mes octubre y dia 14 y 15
eventos=eventos[eventos.mes==10]
eventos=eventos[(eventos.dias==14) | (eventos.dias==15)]
eventos.mes.value_counts()
```

```python
# inner join
combinado1= pd.merge(eventos,locacion, how='inner', on='id_locacion')
```

```python
#selecciono columnas de archivo combinado1
agrupo=combinado1[['sede','cant_espectadores']]
agrupo.groupby('sede').count().sort_values(by='cant_espectadores', ascending=False).head(1)
```

#### b) Nombre del evento y nombre de la categoria deportiva de aquellos eventos cuya cantidad de espectadores supero la capacidad de la locacion, mas alla de la capacidad extendida. Esto es vital importancia para detectar problemas de seguridado si es necesario realizar algun cambio de locacion

```python
#forma de la sugerencia de la profesora
combinado2=pd.merge(eventos,locacion,how='inner', on='id_locacion')
combinado2['capacidad_maxima'] = combinado2.capacidad + combinado2.capacidad_extendida
combinado2['supero_capacidad'] = combinado2.capacidad > combinado2.capacidad_maxima
combinado2
```

```python
combinado3 = pd.merge(combinado2,categorias_deportivas, how='inner', on='id_categorias')
combinado3 = combinado3.groupby(['nombre_evento','nombre_categoria']).groups
combinado3

```
